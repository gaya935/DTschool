### 딕셔너리 실습_편의점 재고 관리 

```python
#물건이름과 갯수 쌍을 이룸 ->엔터를 입력할 때까지 반복
#편의점 재고 정리
goods = {}
while True:
    product = input('물건이름을 넣어 주세요.: ')
    if product =='':
        break
    number = int(input('갯수: ')) 
    #goods[product] = number

    if product in goods.keys():
        goods[product] += number
    else:
        goods[product] = number
print(goods)
# 오름차순 정렬
for key, value in sorted(goods.items()):
    print(key, value, end=' ')
    
```

```python
#민준님 코드
price = int(input('물건의 소매자가격을 입력하세요: '))
cut = 0
actual = 0
if price >= 100000 :
    cut = 10
elif price >= 50000 : 
    cut = 7.5
elif price >= 10000 : 
    cut = 5
else: 
    cut = 0
actual = round(price - price/100*cut) # round추가
dis_price = round( price/100*cut)

print(f'소매자가격: {price}원')
print(f'할인율: {cut} %')
print(f'할인금액: {dis_price}원')
print(f'할인된 가격: {actual}원')
```

```python
#가연님 코드 
price=int(input('물건 구매가를 입력하세요: '))

if price>100000:
    discount=price*0.1
elif price>50000:
    discount=price*0.075
elif price>10000:
    discount=price*0.05

print(f'구매금액: {price}원')
print(f'할인율: {(discount/price)*100}%') #*100추가
print(f'할인금액: {int(discount)}원')
print(f'지불금액: {int(price-discount)}원')
```

```python
#사칙연산 프로그램
print('나눗셈 연산시', '0','은 입력하지 마세요.')
num1 = int(input('숫자1: '))
num2 = int(input('숫자2: '))
calculator = input('+,-,/,* : ')

if calculator == '+':
    print(num1+num2)
elif calculator == '*':
    print(num1*num2)
elif calculator == '-':
    print(num1-num2)
elif calculator == '/':
    
    if num1 == 0 or num2 == 0:
        print ('숫자를 다시 입력하세요.')
    else:
        print(float(num1/num2))
```

```python
#사칙연산 프로그램2
print('나눗셈 연산시', '0','은 입력하지 마세요.')
num1 = int(input('숫자1: '))
num2 = int(input('숫자2: '))
calculator = input('+,-,/,* : ')
result = 0

if calculator == '+':
    result = num1+num2
elif calculator == '*':
    result = num1*num2
elif calculator == '-':
    result = num1-num2
elif calculator == '/':
    
    if num1 == 0 or num2 == 0: # num2 !=0:아니면 계산으로 넘어가는 코드도 됨.
        print('숫자를 다시 입력하세요.')
    else:
        result = round(float(num1/num2),2)
print('-'*30)
print(f'{num1}{calculator}{num2} = {result}')
```

```python
#주사위 던지기

count1 = 0
count2 = 0

for times in range(10):
    dice1 = random.randint(1,6)
    dice2 = random.randint(1,6)
    result = 0
    winner = 0  
    if dice1 > dice2 :
        result = '이겼습니다.'
        winner = 1
        count1 +=1
    elif dice1 < dice2:
        result = '이겼습니다.'
        winner = 2
        count2 +=1
    else: 
        result = '비겼습니다.'
        winner = '1과 주사위2가'
    print(f'{times}번째판은 주사위{winner}이(가) {result} ')

f_winner = 0
f_result = 0
if count1 > count2:
    f_winner = '주사위1'
    f_result = '이겼습니다.'
elif count2 > count1:
    f_winner = '주사위2'
    f_result = '이겼습니다.' 
else:
    f_winner = '둘 다'
    f_result = '비겼습니다.'

print('-'*30)
print(f'주사위1:주사위2 = {count1}: {count2}로 {f_winner}이(가) {f_result}')
```
